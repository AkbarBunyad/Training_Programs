```python
import pandas as pd
path = '/content/drive/MyDrive/Colab Notebooks/cars.xlsx'
df = pd.read_excel(path)
df.head()
```

1. What is the average city MPG (cty) across all cars in the dataset?

```python
df['cty'].mean()
```

2. Calculate the average MPG (both city and highway) for each car class

```python
df.groupby(by = 'class').agg({'cty': 'mean', 'hwy': 'mean'})
```

3. What is the distribution of cars by transmission type (trans)?

```python
df['trans'].value_counts()
```

4. List the top 3 most fuel-efficient cars (highest average of city and highway MPG) for each fuel type (fl)

```python
df['av_cty_hwy'] = 0.5 * (df['cty'] + df['hwy'])
df.sort_values(by = 'av_cty_hwy', ascending = False).groupby('fl').head(3)
```

5. How many unique manufacturers are present in the dataset

```python
df['manufacturer'].nunique()
```

6. Which year has the most car models listed in the dataset?

```python
df['year'].value_counts().idxmax()
```

7. For each manufacturer, find the model with the highest highway MPG (hwy).

```python
idx = df.groupby('manufacturer')['hwy'].idxmax()
df.loc[idx, ['manufacturer', 'model', 'hwy']]
```

8. Compare the average highway MPG of cars with automatic and manual transmissions

```python
df['new_category'] = df['trans'].apply(lambda x: 'auto' if 'auto' in x else 'manual')
df.groupby('new_category')['hwy'].mean()
```

9. What is the average displacement (displ) of cars for each number of cylinders (cyl)?

```python
df.groupby('cyl')['displ'].mean()
```

10. For cars produced after 2005, how does the average highway MPG compare across different drive types (drv)?

```python
df[df['year'] > 2005].groupby('drv')['hwy'].mean()
```

11.Identify the manufacturer and model of cars that have a city MPG above the 75th percentile and a highway MPG below the 50th percentile.

```python
mpg = df['cty'].quantile(0.75)
hwy = df['hwy'].quantile(0.5)

new_df = df[(df['cty'] > mpg) & (df['hwy'] < hwy)]
new_df
```

12. Calculate the variance and standard deviation of highway MPG for each car class

```python
df.groupby('class')['hwy'].agg(['std', 'var'])
```

13. Which manufacturer has the widest range of highway MPG across their models, and what is that range?

```python
new_df = df.groupby('manufacturer')['hwy'].std()
nu = new_df.idxmax()
num = new_df.max()
print(nu)
print(num)
```
